```python
sns.pairplot(df.drop("CustomerID",axis=1))
```

```python
# segment customers by gender
sns.countplot(x='Gender', data=df)
plt.title('Customer Gender Distribution')
plt.show()
```

Through above analysis we understood that the number of female customers are more than number of male customers in the mall.

```python
sns.histplot(x="Age",data=df,color='blue',bins=20)
plt.title("Age Distribution")
plt.show()
```

From above analysis we understood that the age group of customers are highly between 30-35 years. The lower age group are  between 70-75 years.

```python
sns.kdeplot(x='Annual Income ($)', data=df, color="green", fill=True)
plt.title('Income Distribution')
plt.show()
```

The  majority of customers have annual income  betweeen 70k$ - 90k$. The lesser customers have annual income  more than 20k$

```python
sns.kdeplot(x='Work Experience',data=df,color='blue',fill=True)
plt.title("Work Experience Distribution")
plt.show()
```

The major work experience of customers lies between 0-3 years.Lesser customers have Work experience between 15-20 years.

```python
sns.countplot(x='Family Size',data = df)
plt.title('Customer Family Size Distribution')
plt.show()
```

Majority of the customers have family size of 2 members whereas there are less customers with the family size of 8-9 members

```python
sns.countplot(x='Profession',data=df)
plt.xticks(rotation=35)
plt.title("Distribution on basis of Profession")
plt.show()
```

The most of the customers have profession: Artist and less number of customers belong to Homemaker profession.

```python
sns.violinplot(x='Gender',y='Spending Score (1-100)',data=df)
plt.title("Spending Score based on Gender")
plt.show()
```

Through violin plot we analyze that Male and Female have similar spending behaviour

```python
sns.lineplot(x='Age',y='Spending Score (1-100)',data=df)
plt.title("Spending score based on Age")
plt.show()
```

```python
plt.hexbin(x="Annual Income ($)",y="Spending Score (1-100)",data=df,gridsize=20,cmap='Blues')
plt.xlabel('Annual Income ($)')
plt.xticks(rotation=45)
plt.ylabel('Spending_Score')
plt.title("Spending Score by Income")
plt.colorbar()
plt.show()
```

Highly spending behaviour is seen after 50k$ annual income and lesser in below 50k$ annual income

```python
sns.barplot(x='Profession',y='Spending Score (1-100)',data=df) 
plt.xticks(rotation=45)
plt.title('Spending Score by Profession')
plt.show()
```

The spending behaviour of profession Entertainment is the highest whereas the Homemaker have the lowest spending behaviour

```python
sns.boxplot(x='Work Experience',y='Spending Score (1-100)',data=df) #doubt
plt.title("Spending Score by Work Experience")
plt.show()
```

Spending behaviour of people having work experience 2-3,7 years are highest whereas people having work experience 16 years have lowest

```python
sns.violinplot(x='Family Size',y='Spending Score (1-100)',data=df)
plt.title("Spending Score by Family Size")
plt.show()
```

The violin plot above shows people having family size -8 have lesser spending behaviour than people having family size - 2,3
